## Nettoyage de ___products___

```python
products = pd.read_csv("data/endogene/products_export_1.csv")
```

```python
products.columns.tolist()
```

```python
products = products.get([ 
    'Title',
    'Custom Product Type',
    'Status'
])
```

```python
products
```

```python
products = products.dropna(subset=['Title'])
```

```python
products.shape
```

```python
# len(products.Title.unique().tolist())
products.nunique()
```

```python
products.drop_duplicates(subset="Title", keep = 'first', inplace=True)
```

```python
products
```

```python
products.columns = [ 
    'title',
    'category',
    'status'
]
```

```python
products
```

```python
products.to_csv("data/endogene/clean data/products.csv")
```
