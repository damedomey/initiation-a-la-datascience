```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha = 'center')
```

```python
yes = customers.accepts_marketing[customers.accepts_marketing == 'yes'].count()
accept_marketing = [yes, customers.accepts_marketing.count() - yes]
plt.bar(['yes','no'], accept_marketing, color=('green','red'))
plt.yticks(accept_marketing)
plt.title("Acceptation du marketing")
plt.savefig("data/images/Acceptation du marketing.jpg")
```

```python
# customers = customers.astype({'orders': 'str'})
x = customers.orders.sort_values().unique()
x = x.astype('str')
y = [customers.orders[customers.orders == int(i)].count() for i in x]
plt.bar(x,y)
plt.xlabel('orders')
plt.ylabel('customer')
addlabels(x,y)
plt.title("Statistique par nombre de commandes")
plt.savefig("data/images/Statistique par nombre de commandes.jpg")
```

```python
sales = sales[sales.quantity >= 0]
# Catégorie de produit la plus vendue
category = sales.product_category.unique().tolist()
sumSales = [sales.quantity[sales.product_category == c].sum() for c in category]
sumOrders = [orders.quantity[orders.category == c].sum() for c in category]
plt.figure(figsize=(15,10))
plt.plot(category, sumSales)
plt.plot(category, sumOrders)
plt.tick_params(axis='x', rotation=90,)
plt.legend(('Sales', 'Orders'))
plt.xlabel('Catégorie de produit')
plt.ylabel('Quantité')
plt.title("Statistiques de sales et orders par catégorie")
plt.savefig("data/images/Statistiques de sales et orders par catégorie.jpg", bbox_inches='tight')
```
