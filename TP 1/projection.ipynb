```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i], ha = 'center')
```

```python
yes = customers.accepts_marketing[customers.accepts_marketing == 'yes'].count()
accept_marketing = [yes, customers.accepts_marketing.count() - yes]
plt.bar(['yes','no'], accept_marketing, color=('green','red'))
plt.yticks(accept_marketing)
plt.title("Acceptation du marketing")
```

```python
# customers = customers.astype({'orders': 'str'})
x = customers.orders.sort_values().unique()
x = x.astype('str')
y = [customers.orders[customers.orders == int(i)].count() for i in x]
plt.bar(x,y)
plt.xlabel('orders')
plt.ylabel('customer')
addlabels(x,y)
plt.title("Statistique par nombre de commandes")
```
